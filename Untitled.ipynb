```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# For load the data
df = pd.read_csv("retail_discount_data.csv.txt", delimiter="\t")
df.to_csv("retail_discount_data.csv",index=False)
```

```python
df.head()
```

```python
# For representing the discount applied data
discount_df = df[df["Discount_Applied"] == "Yes"]
```

```python
discount_df.head()
```

```python
 df.columns
```

```python
# Count the customers use discount on the particular date
daily_discount_count = discount_df.groupby("Date")["Customer_ID"].count().reset_index()
```

```python
# Rename columns for better readability
daily_discount_count.columns = ["Date","Discount_Usage_Count"]
```

```python
daily_discount_count.head()
```

```python
#visualizations
plt.figure(figsize=(10,5))
plt.plot(daily_discount_count["Date"],daily_discount_count["Discount_Usage_Count"],marker='o', color='blue')
plt.title("Daily Discount Usage Trend")
plt.xlabel("Date")
plt.ylabel("Number of Discounts Used")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Discount_Percentage"], df["Purchase_Amount"], color='green')
plt.title("Discount Percentage vs Purchase Amount")
plt.xlabel("Discount Percentage")
plt.ylabel("Purchase Amount")
plt.grid(True)
plt.show()
```

```python
# Affect customer purchase amount
correlation = df["Discount_Percentage"].corr(df["Purchase_Amount"])
print("Correlation between Discount Percentage and Purchase Amount:", correlation)
```

```python
# Function to categorise discount
def discount_category(x):
    if x == 0:
        return "No Discount"
    elif x <= 10:
        return "Low Discount"
    elif x <= 20:
        return "Medium Discount"
    else:
        return "High Discount"

# Apply the function to create a new column
df["Discount_Category"] = df["Discount_Percentage"].apply(discount_category)

df.head()
```
